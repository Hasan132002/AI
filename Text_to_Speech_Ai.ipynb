```python
def text_to_speech(text,language='en'):
  tts=gTTS(text=text,lang=language)
  tts.save("voice_one.mp3")
  return Audio("voice_one.mp3",autoplay=True)
#  os.system("start voice_one.mp3")
  # print('Audio has been generated')

def read_text_from_file(file_path):
  with open(file_path,'r') as file:
    text=file.read()
  return text

file_path="intro.txt"
text=read_text_from_file(file_path)

#text="Hello I have done such an Insane things"
audio=text_to_speech(text)
display(audio)
```

```python
import pyttsx3
from pyttsx3.drivers import espeak

def text_to_speech(text, voice='male', output_file='output.mp3'):
    engine = pyttsx3.init()
    voices = engine.getProperty('voices')
    if voice == 'male':
        for v in voices:
            if "male" in v.name.lower():
                engine.setProperty('voice', v.id)
                break
    else:
        for v in voices:
            if "female" in v.name.lower():
                engine.setProperty('voice', v.id)
                break
    engine.save_to_file(text, output_file)
    engine.runAndWait()

def read_text_from_file(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
    return text

file_path = "intro.txt"
text = read_text_from_file(file_path)
text_to_speech(text, voice='male')
```
