ML Model 1: Linear Regression for Time Series Forecasting
(Predict unemployment for Pakistan till 2024)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

file_path = "/content/long_format_unemployment.csv"
df = pd.read_csv(file_path)

pak = df[df['Country'] == 'Pakistan'].sort_values(by='Year')
print(pak)
```

```python
years = pak['Year'].values.reshape(-1, 1)
values = pak['Unemployment'].values.reshape(-1, 1)

mask = ~np.isnan(values.flatten())
years_clean = years[mask]
values_clean = values[mask]

model = LinearRegression()
model.fit(years_clean, values_clean)

future_years = np.array(range(2021, 2028)).reshape(-1, 1)
future_preds = model.predict(future_years)

plt.figure(figsize=(10, 5))
plt.plot(years_clean, values_clean, label='Actual', marker='o')
plt.plot(future_years, future_preds, '--', label='Prediction', marker='x')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Prediction for Pakistan')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

for year, pred in zip(future_years.flatten(), future_preds.flatten()):
    print(f"Predicted Unemployment in {year}: {pred:.2f}%")
```

**Insights from the Model**
The model was trained on historical unemployment data for Pakistan from 2000 to 2023.

We then predicted unemployment for the years 2021 to 2027.

A line plot was created showing both the actual data and predicted values to help visualize the trend.

If the historical data shows a steady increase or decrease, the predicted values will follow that trend.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Load the dataset
file_path = "/content/long_format_unemployment.csv"
df = pd.read_csv(file_path)

# Filter for Pakistan
pakistan_df = df[df['Country'] == 'Pakistan']

# Use data from 2010 onwards to train model
pakistan_df = pakistan_df[pakistan_df['Year'] >= 2010]

# Prepare data
X = pakistan_df[['Year']].values
y = pakistan_df['Unemployment'].values

# Train Linear Regression
model = LinearRegression()
model.fit(X, y)

# Predict for years 2019 to 2024
future_years = np.array(range(2019, 2025)).reshape(-1, 1)
predicted_unemployment = model.predict(future_years)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(pakistan_df['Year'], pakistan_df['Unemployment'], label='Actual', marker='o')
plt.plot(future_years, predicted_unemployment, '--', label='Predicted (2019–2024)', marker='x')

# Annotate actual values
for x, y_val in zip(pakistan_df['Year'], pakistan_df['Unemployment']):
    plt.text(x, y_val + 0.2, f'{y_val:.2f}', ha='center', fontsize=8)

# Annotate predicted values
for x, y_pred in zip(future_years.flatten(), predicted_unemployment):
    plt.text(x, y_pred + 0.2, f'{y_pred:.2f}', ha='center', fontsize=8, color='red')

plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate in Pakistan (2019–2024)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

file_path = "/content/long_format_unemployment.csv"
df = pd.read_csv(file_path)

```

# **#  Clean missing values**

```python
df.dropna(subset=["Year", "Unemployment"], inplace=True)
df['Region'] = df['Region'].fillna("Unknown")
df['IncomeGroup'] = df['IncomeGroup'].fillna("Unknown")
```

# **#Trend Analysis: Global Unemployment Over Time**


```python
global_trend = df.groupby('Year')['Unemployment'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=global_trend, x='Year', y='Unemployment', marker='o')
plt.axvspan(2008, 2009, color='red', alpha=0.2, label='2008 Crisis')
plt.axvspan(2020, 2022, color='orange', alpha=0.2, label='COVID-19')
plt.title("📉 Global Unemployment Trend (1991–2024)")
plt.xlabel("Year")
plt.ylabel("Avg Unemployment Rate (%)")
plt.grid(True)
plt.legend()
plt.show()
```

# **Regional Comparisons: Highest & Lowest Averages**



```python
region_avg = df.groupby('Region')['Unemployment'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=region_avg.values, y=region_avg.index)
plt.title("🌍 Average Unemployment by Region")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Region")
plt.show()
```

# **Income Group Comparisons**


```python

income_avg = df.groupby('IncomeGroup')['Unemployment'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(x=income_avg.values, y=income_avg.index, palette="coolwarm")
plt.title("💰 Unemployment Rate by Income Group")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Income Group")
plt.show()
```
