```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the files
file1_url = "https://drive.google.com/uc?export=download&id=1Vq38NInjeZ-sG-ndA7_sLPe3Bws1o3x8"
file2_url = "https://drive.google.com/uc?export=download&id=1Vq38NInjeZ-sG-ndA7_sLPe3Bws1o3x8"
#file2_url = "https://drive.google.com/uc?export=download&id=1CpaT6WHgYlX_sHvQe0-0wgJGmn3RjLkH"

file1 = pd.read_csv(file1_url)
file2 = pd.read_csv(file2_url)

# Prepare the data for semantic similarity comparison
sentences1 = file1['sentence'].tolist()
sentences2 = file2['sentence'].tolist()

# Use SentenceTransformer for embedding sentences
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True)

# Calculate cosine similarity between sentence embeddings
similarity_matrix = cosine_similarity(embeddings1, embeddings2)

# Threshold for considering a match
threshold = 0.9

# Count the number of matches above the threshold
matches = sum(similarity_matrix.max(axis=1) > threshold)

print("Number of semantic matches between the two files:", matches)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the files
file1_path = "/content/IMC_Rayan.csv"
file2_path = "/content/KCH_Rayan.csv"

file1 = pd.read_csv(file1_path)
file2 = pd.read_csv(file2_path, encoding='ISO-8859-1')  # Specify the encoding

# Prepare the data for semantic similarity comparison
sentences1 = file1['ItemDescription (IMC)'].tolist()  # Update column name
sentences2 = file2['FINAL DESCRIPTION'].tolist()  # Update column name

# Use SentenceTransformer for embedding sentences
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings1 = model.encode(sentences1, convert_to_tensor=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True)

# Calculate cosine similarity between sentence embeddings
similarity_matrix = cosine_similarity(embeddings1, embeddings2)

# Threshold for considering a match
threshold = 0.9

# Count the number of matches above the threshold
matches = sum(similarity_matrix.max(axis=1) > threshold)

print("Number of semantic matches between the two files:", matches)
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

def load_csv(file_path, encoding='utf-8'):
    try:
        return pd.read_csv(file_path, encoding=encoding)
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"An error occurred while loading the file: {str(e)}")
        return None

def calculate_semantic_similarity(sentences1, sentences2, threshold=0.6):
    model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
    embeddings1 = model.encode(sentences1, convert_to_tensor=True)
    embeddings2 = model.encode(sentences2, convert_to_tensor=True)
    similarity_matrix = cosine_similarity(embeddings1, embeddings2)
    matches = sum(similarity_matrix.max(axis=1) > threshold)
    return matches, similarity_matrix

def visualize_similarity(similarity_matrix):
    plt.imshow(similarity_matrix, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.xlabel('Sentences from file 2')
    plt.ylabel('Sentences from file 1')
    plt.title('Cosine Similarity Matrix')
    plt.show()

file1_path = "/content/IMC_Rayan.csv"
file2_path = "/content/KCH_Rayan.csv"

file1 = load_csv(file1_path)
file2 = load_csv(file2_path, encoding='ISO-8859-1')

if file1 is not None and file2 is not None:
    sentences1 = file1['ItemDescription (IMC)'].tolist()
    sentences2 = file2['FINAL DESCRIPTION'].tolist()

    matches, similarity_matrix = calculate_semantic_similarity(sentences1, sentences2)
    print("Number of semantic matches between the two files:", matches)

    visualize_similarity(similarity_matrix)
```
